```python
import pandas as pd
```

```python
df = pd.read_csv("OLA.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.columns
```

```python
df.duplicated().sum()
```

```python
df['V_TAT'] = pd.to_numeric(df['V_TAT'], errors='coerce').fillna(0)
df['C_TAT'] = pd.to_numeric(df['C_TAT'], errors='coerce').fillna(0)
```

```python
df.isnull().sum()
```

```python
df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna('No')
```

```python
df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna('No')
```

```python
df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna('No')
```

```python
df['Incomplete_Rides_Reason'] = df['Incomplete_Rides_Reason'].fillna('Not Applicable')
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna('Not Applicable')
```

```python
df['Driver_Ratings'] = pd.to_numeric(df['Driver_Ratings'], errors='coerce').fillna(0)
df['Customer_Rating'] = pd.to_numeric(df['Customer_Rating'], errors='coerce').fillna(0)
```

```python
df['Booking_Status'] = df['Booking_Status'].str.strip().str.title()
```

```python
df.isnull().sum()
```

```python
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
df.loc[(df['Driver_Ratings'] < 0) | (df['Driver_Ratings'] > 5), 'Driver_Ratings'] = 0
df.loc[(df['Customer_Rating'] < 0) | (df['Customer_Rating'] > 5), 'Customer_Rating'] = 0
```

```python
df.loc[df['Ride_Distance'] < 0, 'Ride_Distance'] = 0
df.loc[df['Booking_Value'] < 0, 'Booking_Value'] = 0
```

```python
df.duplicated(subset=['Booking_ID']).sum()
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.replace(' ', '_')
)
```

```python
df.info()
df.isnull().sum()
df.head()
```

```python
df.to_csv("ola_clean.csv", index=False)
```

```python
df = pd.read_csv("ola_clean.csv")
```

```python
df['ride_datetime'] = pd.to_datetime(
    df['date'].astype(str) + ' ' + df['time'].astype(str),
    errors='coerce'
)
```

```python
df['ride_hour'] = df['ride_datetime'].dt.hour
```

```python
df['ride_day'] = df['ride_datetime'].dt.day_name()
```

```python
df['ride_month'] = df['ride_datetime'].dt.month
```

```python
df['tat_efficiency'] = df['v_tat'] + df['c_tat']
```

```python
df['ride_distance_bucket'] = pd.cut(
    df['ride_distance'],
    bins=[0, 5, 15, 30, 100],
    labels=['Very Short', 'Short', 'Medium', 'Long']
)
```

```python
df.to_csv("ola_clean.csv", index=False)
```

```python
df['ride_distance_bucket'] = df['ride_distance_bucket'].cat.add_categories(['Not Applicable'])
df['ride_distance_bucket'] = df['ride_distance_bucket'].fillna('Not Applicable')
```

```python
df.drop(columns=['date', 'time'], inplace=True)
```

```python
df.to_csv("ola_clean.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("ola_clean.csv")
```

```python
df.shape
df.info()
df.head()
```

```python
df['booking_status'].value_counts()
```

```python
df['booking_status'].value_counts().plot(kind='bar')
plt.title("Booking Status Distribution")
plt.show()
```

Booking Status Distribution

Most rides are successfully completed, indicating strong demand and operational stability.

Driver-side cancellations are higher than customer cancellations, suggesting driver availability or acceptance issues.

```python
df.groupby('ride_hour').size().plot(kind='line')
plt.title("Ride Volume by Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Rides")
plt.show()
```

Ride demand is relatively consistent throughout the day with peaks during late morning and evening hours.

Slight dips during early morning and late night indicate lower off-peak demand.

```python
df.groupby('ride_day').size().plot(kind='bar')
plt.title("Ride Volume by Day")
plt.show()
```

Tuesday and Monday show the highest ride volumes, reflecting strong weekday demand.

Weekend rides are slightly lower, indicating commute-driven usage patterns.

```python
df['vehicle_type'].value_counts().plot(kind='bar')
plt.title("Rides by Vehicle Type")
plt.show()
```

Ride demand is evenly distributed across vehicle types, showing diversified customer preferences.

Premium and economy vehicles both contribute significantly, indicating balanced market demand.

```python
df['ride_distance_bucket'].value_counts().plot(kind='bar')
plt.title("Ride Distance Buckets")
plt.show()
```

A large portion of rides fall under â€œNot Applicableâ€, mainly due to cancellations and failed bookings.

Among completed rides, long and medium-distance trips dominate, contributing more to revenue.

```python
sns.histplot(df[df['driver_ratings'] > 0]['driver_ratings'])
plt.title("Driver Ratings Distribution")
plt.show()
```

Driver ratings are consistently high, mostly ranging between 3.5 to 5, indicating good service quality.

Very few low ratings suggest strong driver performance across the platform.

```python
sns.histplot(df[df['customer_rating'] > 0]['customer_rating'])
plt.title("Customer Ratings Distribution")
plt.show()
```

Customers also show high satisfaction levels, with ratings concentrated above 3.5.

The balanced distribution reflects mutual accountability between riders and drivers.

```python
sns.scatterplot(x='ride_distance', y='booking_value', data=df)
plt.title("Ride Distance vs Booking Value")
plt.show()
```
